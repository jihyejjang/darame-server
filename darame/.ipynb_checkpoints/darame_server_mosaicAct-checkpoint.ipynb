```python
#darame   
#1: mosaic   , composite /   (android->python)
#2: image instance   (python -> android),   
#3:     (android->python)

from socket import *
import socket
import os
import time
import sys
import base64
import io
from PIL import Image
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import darame_instance_segmentation
import darame_mosaic_selected


HOST = "172.16.24.239" # Symbolic name meaning all available interfaces
PORT = 9999 # Arbitrary non-privileged port
 
BACKLOG = 10

img_path = "C:/Users/user/  /CNN-PhotoEditor-App/darame/img/"


#  
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind((HOST, PORT))
server_socket.listen(BACKLOG)
print("TCPServer Waiting for client on port 9999")


#    .
    
client_socket, address = server_socket.accept()
    #   
print("I got a connection from ", address)

data = None

# Data 

recv_data = client_socket.recv(1024)
data = recv_data


while (recv_data):
    recv_data = client_socket.recv(1024)
    data += recv_data
    #print(recv_data)
        
#   

print("finish img recv")
print(sys.getsizeof(data))
#print(data)
    
#i = base64.b64decode(data)
i = io.BytesIO(data)
i = mpimg.imread(i, format='JPG')

plt.imshow(i, interpolation='nearest')
plt.show()

Image.fromarray(np.array(i).astype(np.uint8)).save(img_path + "mosaic.png")

# 
#darame_instance_segmentation.main(mosaic=True) #input 'img/mosaic.png', output 'output/display_maskes.png'
# , X,Y 

f=open("output/display_maskes.png",'rb')# open file as binary

data=f.read()

client_socket.sendall(data)

#print (data)

f.flush()

f.close()

#X,Y  ,  
X=200
Y=300 #
#darame_mosaic_selected.main(X,Y) # output 'output/display_result_of_mosaic.png'

#Todo:        

print("Finished ")

client_socket.close()
print("SOCKET closed... END")
```
